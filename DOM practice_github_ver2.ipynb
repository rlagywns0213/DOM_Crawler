### 사이트 자체에서 막아놓은 부분은 크롤링하기 어려울때 유용한 방법이 될수도 있음 
### 그러나, 일일이 html문을 로컬에 저장하면서 실행하는 방식은 매우 비효율적으로 판단됨

### 크롤링 코드를 구현하다 에러가 나면 pass 하는 형식으로 코드를 짠후, 위의 라이브러리를 이용하여 페이지 하나하나를 해주면 어떨까는 생각이 듬

# 활용 예시

```python
def website_input(web,content,content2):
    import requests
    from bs4 import BeautifulSoup
    from body_text_extraction import BodyTextExtraction
    
    base_url = web
    resp = requests.get(base_url)
    html_text = resp.text
    bte = BodyTextExtraction()
    text = bte.extract(html_text)
    text_fixed = text.replace('\t', '').replace('\n', '')
    #첫번쨰 인덱스
    content_index = text_fixed.find(content)
    #두번째 인덱스
    next_content_index = text_fixed.find(content2)
    real_text = text_fixed[content_index:next_content_index]
    
    return real_text
```

```python
website_input('https://www.data.go.kr/data/15075578/fileData.do','파일데이터명','분류체계')
```

```python
website_input('https://www.data.go.kr/data/15075578/fileData.do','제공기관','관리부서명')
```
