# test2 - input값이 web인 함수로 생성 

```python
def website_input(web):
    import requests
    from bs4 import BeautifulSoup
    from cetd.extractor import Extractor
    
    base_url = web
    resp = requests.get(base_url)
    html_text = resp.text
    
    ext = Extractor()
    extracted = ext.extract_content(html_text)
    
    return extracted
```

```python
website_input('https://finance.naver.com/news/news_read.nhn?article_id=0004520246&office_id=008&mode=mainnews&type=&date=2020-12-28&page=1')
```
